```python
model = tf.keras.models.Sequential()
import math
import random
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
import pandas_datareader as web
dataset = web.DataReader("AXISBANK.NS",data_source='yahoo',start='2012-05-07',end='2020-05-11')
```

```python
import pandas as pd
train_set = dataset['Low']
train_set = pd.DataFrame(train_set)
```

```python
train_set
```

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler(feature_range=(0,1))
strain_set = sc.fit_transform(train_set)
```

```python
strain_set
```

```python
Xtrain = []
Ytrain = []
for i in range(30,1966):
  Xtrain.append(strain_set[i-30:i, 0])
  Ytrain.append(strain_set[i, 0])

Xtrain = np.array(Xtrain)
Ytrain = np.array(Ytrain)
print(Xtrain)
```

```python
Xtrain = np.reshape(Xtrain, (Xtrain.shape[0], Xtrain.shape[1], 1) )
```

```python
from tensorflow.keras.layers import LSTM,Dropout,Dense
```

```python
model.add(LSTM(units = 50, return_sequences = True, input_shape = (30, 1)))
model.add(Dropout(0.2))
model.add(LSTM(units = 50, return_sequences = True))
model.add(Dropout(0.2))
model.add(LSTM(units = 50, return_sequences = True))
model.add(Dropout(0.2))
model.add(LSTM(units = 50, return_sequences = True))
model.add(Dropout(0.2))
model.add(LSTM(units = 50, return_sequences = False))
model.add(Dropout(0.2))
model.add(Dense(units = 1
               ))
```

```python
model.compile(optimizer = 'adam', loss = 'mean_squared_error')
```

```python
model.fit(Xtrain, Ytrain, epochs = 150, batch_size = 32)
```
